## Preprocessing data

### Subtask:
Lakukan preprocessing data, termasuk menangani nilai yang hilang (jika ada), mengkodekan fitur kategorikal (jika ada), dan membagi data menjadi set pelatihan dan pengujian.


**Reasoning**:
Handle missing values, identify and encode categorical columns, and split the data into training and testing sets.



```python
# Identify and encode categorical columns, excluding 'model_path' and 'model_type' (as it seems to be the target)
object_cols = df_steganalysis.select_dtypes(include='object').columns
categorical_cols_to_encode = [col for col in object_cols if col not in ['model_path', 'model_type']]

print("\nCategorical columns to encode:", categorical_cols_to_encode)

# We will use one-hot encoding for the identified categorical columns.
df_steganalysis_encoded = pd.get_dummies(df_steganalysis, columns=categorical_cols_to_encode, drop_first=True)

# Separate features (X) and target variable (y)
# Assuming the target variable is 'model_type' and 'injected' corresponds to 'tersusupi' (1) and other values to 'clean' (0).
y = df_steganalysis_encoded['model_type'].apply(lambda x: 1 if 'injected' in x else 0)
X = df_steganalysis_encoded.drop(['model_type', 'model_path'], axis=1) # Drop 'model_type' and 'model_path'

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=36, stratify=y)

print("\nShape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Pilih dan latih model klasifikasi

### Subtask:
Pilih model klasifikasi yang sesuai (misalnya, Logistic Regression, Support Vector Machine, Random Forest, atau Neural Network) dan latih model menggunakan data pelatihan.


**Reasoning**:
Import the necessary classification model class and train the model using the training data.



```python
from sklearn.ensemble import RandomForestClassifier

# Create an instance of the Random Forest Classifier model
model = RandomForestClassifier(
    n_estimators=20,
    min_samples_split=2,
    criterion='entropy',
    oob_score=True,
    random_state=36)

# Train the model using the training data
model.fit(X_train, y_train)
```

## Evaluasi model

### Subtask:
Evaluasi kinerja model menggunakan metrik yang relevan seperti akurasi, presisi, recall, dan F1-score pada data pengujian.


**Reasoning**:
Import the necessary evaluation metrics and make predictions on the test data.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Make predictions on the test data
y_pred = model.predict(X_test)
```

**Reasoning**:
Calculate and print the evaluation metrics (accuracy, precision, recall, F1-score) and the classification report.



```python
# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the evaluation metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")

# Print the classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

### Subtask:

**Reasoning**:
Based on the evaluation metrics (accuracy 0.8125, F1-score 0.88), the current model performance is reasonable but there might be room for improvement, especially considering the potential imbalance or the need for higher precision or recall depending on the specific application. To potentially improve the model, I will use GridSearchCV to tune the hyperparameters of the Random Forest Classifier. I will define a parameter grid with a few key hyperparameters to explore.



```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, StratifiedKFold


# Define the parameter grid for GridSearchCV
# param_grid = {
#     'n_estimators': [50, 100, 200],
#     'max_depth': [None, 10, 20],
#     'min_samples_split': [2, 5, 10],
#     'min_samples_leaf': [1, 2, 4]
# }
param_grid = {
    'n_estimators': [10, 15, 20, 50],
    'max_depth':  [5, 10, 15],
    'min_samples_split': [5, 7, 10],
    'min_samples_leaf':  [2, 4, 8],
    'random_state': [28, 36, 42]
}
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=36)

# Create a GridSearchCV object
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(criterion='entropy'),
    param_grid=param_grid,
    cv=cv, # Using 3-fold cross-validation
    scoring='f1', # Optimize for F1-score
    #    n_jobs=-1, # Use all available cores
    ) 

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best parameters found:", best_params)
print("Best F1-score found during cross-validation:", best_score)
```

**Reasoning**:
I have performed hyperparameter tuning using GridSearchCV and found the best parameters. Now, I need to train a new Random Forest model using these best parameters and then evaluate its performance on the test set to see if the tuning improved the model's performance compared to the initial model.



```python
# Train a new model with the best parameters
tuned_model = RandomForestClassifier(**best_params)
tuned_model.fit(X_train, y_train)

# Make predictions on the test data using the tuned model
y_pred_tuned = tuned_model.predict(X_test)

# Evaluate the tuned model
accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
precision_tuned = precision_score(y_test, y_pred_tuned)
recall_tuned = recall_score(y_test, y_pred_tuned)
f1_tuned = f1_score(y_test, y_pred_tuned)

print("\nTuned Model Evaluation:")
print(f"Accuracy: {accuracy_tuned:.4f}")
print(f"Precision: {precision_tuned:.4f}")
print(f"Recall: {recall_tuned:.4f}")
print(f"F1-score: {f1_tuned:.4f}")

print("\nClassification Report (Tuned Model):")
print(classification_report(y_test, y_pred_tuned, zero_division=0))
```

# Task
Train a PyTorch model for 20 epochs to detect whether a model is clean or infected, using the dataset from "/content/steganalysis_features_20251020_113232.csv". The dataset contains features extracted from ResNet cover and stego models, where stego models have malware embedded in their weights by converting float weights to bits, inserting malware bits, and reconstructing back to float. The malware insertion varies across the 1st, 2nd, and 3rd mantissa bits.

## Persiapan data untuk pytorch

### Subtask:
Ubah data dari pandas DataFrames menjadi tensor PyTorch dan buat Dataset dan DataLoader untuk pelatihan dan pengujian.


**Reasoning**:
Convert the pandas DataFrames to PyTorch tensors and create custom Dataset and DataLoader classes for training and testing, then create instances of the Dataset and DataLoader.



```python
import torch
from torch.utils.data import Dataset, DataLoader
# Identify and remove object columns from X_train and X_test
object_cols_in_X_train = X_train.select_dtypes(include='object').columns
if len(object_cols_in_X_train) > 0:
    print(f"Object columns found in X_train: {list(object_cols_in_X_train)}")
    X_train_cleaned = X_train.drop(columns=object_cols_in_X_train)
    X_test_cleaned = X_test.drop(columns=object_cols_in_X_train)
else:
    X_train_cleaned = X_train.copy()
    X_test_cleaned = X_test.copy()
    print("No object columns found in X_train.")
# Convert cleaned pandas DataFrames to NumPy arrays with explicit float type
X_train_np = X_train_cleaned.values.astype('float32')
X_test_np = X_test_cleaned.values.astype('float32')

# Convert NumPy arrays to PyTorch tensors
X_train_tensor = torch.Tensor(X_train_np)
y_train_tensor = torch.Tensor(y_train.values).long()
X_test_tensor = torch.Tensor(X_test_np)
y_test_tensor = torch.Tensor(y_test.values).long()

# Create a custom Dataset class
class SteganalysisDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]

# Create instances of the custom Dataset
train_dataset = SteganalysisDataset(X_train_tensor, y_train_tensor)
test_dataset = SteganalysisDataset(X_test_tensor, y_test_tensor)

# Create instances of DataLoader
batch_size = 3 # Define batch size
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print("\nTrain dataset size:", len(train_dataset))
print("Test dataset size:", len(test_dataset))
print("Number of batches in train_dataloader:", len(train_dataloader))
print("Number of batches in test_dataloader:", len(test_dataloader))
```

## Definisikan model pytorch

### Subtask:
Definisikan arsitektur model detektor menggunakan PyTorch. Ini bisa berupa model yang sudah ada atau arsitektur kustom.


**Reasoning**:
Import the necessary PyTorch modules and define the custom neural network architecture.



```python
import torch
import torch.nn as nn

# Define the custom neural network class
class SteganalysisDetector(nn.Module):
    def __init__(self, input_dim):
        super(SteganalysisDetector, self).__init__()
        # Define the layers of the neural network
        self.fc1 = nn.Linear(input_dim, 128) # First fully connected layer
        self.relu = nn.ReLU() # ReLU activation function
        self.fc2 = nn.Linear(128, 64) # Second fully connected layer
        self.fc3 = nn.Linear(64, 2) # Output layer (2 classes: clean, tersusupi)

    def forward(self, x):
        # Define the forward pass
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        return x

# Get the input dimension from the training data tensor
input_dim = X_train_tensor.shape[1]

# Create an instance of the custom neural network model
model_pytorch = SteganalysisDetector(input_dim)

print("PyTorch model architecture defined.")
print(model_pytorch)
```

## Definisikan loss function dan optimizer

### Subtask:
Tentukan loss function yang sesuai untuk tugas deteksi dan optimizer untuk memperbarui bobot model.


**Reasoning**:
Define the loss function and the optimizer for the PyTorch model.



```python
import torch.optim as optim

# Define the loss function (Cross-Entropy Loss for classification)
criterion = nn.CrossEntropyLoss()

# Define the optimizer (Adam optimizer)
optimizer = optim.Adam(model_pytorch.parameters(), lr=0.00001)

print("Loss function (CrossEntropyLoss) defined.")
print("Optimizer (Adam) defined with learning rate 0.001.")
```

## Latih model

### Subtask:
Implementasikan loop pelatihan untuk melatih model selama 20 epoch, termasuk forward pass, perhitungan loss, backward pass, dan langkah optimizer.


**Reasoning**:
Implement the training loop for 20 epochs, including forward pass, loss calculation, backward pass, optimizer step, and zeroing gradients.



```python
# Set the model to training mode
model_pytorch.train()

# Training loop
num_epochs = 200
for epoch in range(num_epochs):
    running_loss = 0.0
    for inputs, labels in train_dataloader:
        # Zero the gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model_pytorch(inputs)

        # Calculate the loss
        loss = criterion(outputs, labels)

        # Backward pass
        loss.backward()

        # Update the weights
        optimizer.step()

        # Accumulate the loss
        running_loss += loss.item()

    # Print the average loss for the epoch
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(train_dataloader):.4f}")

print("\nTraining finished.")
```

## Evaluasi model (pytorch)

### Subtask:
Evaluasi kinerja model PyTorch yang telah dilatih pada data pengujian menggunakan metrik yang relevan untuk tugas deteksi.


**Reasoning**:
Set the model to evaluation mode, disable gradient calculation, iterate through the test dataloader, make predictions, collect true and predicted labels, and then calculate and print the evaluation metrics and classification report.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import numpy as np

# Set the model to evaluation mode
model_pytorch.eval()

# Disable gradient calculation for inference
with torch.no_grad():
    all_labels = []
    all_predictions = []
    for inputs, labels in test_dataloader:
        # Forward pass
        outputs = model_pytorch(inputs)

        # Get predicted class labels
        _, predicted = torch.max(outputs, 1)

        # Store true and predicted labels
        all_labels.extend(labels.numpy())
        all_predictions.extend(predicted.numpy())

# Convert lists to numpy arrays for evaluation
all_labels = np.array(all_labels)
all_predictions = np.array(all_predictions)

# Calculate evaluation metrics
accuracy = accuracy_score(all_labels, all_predictions)
precision = precision_score(all_labels, all_predictions)
recall = recall_score(all_labels, all_predictions)
f1 = f1_score(all_labels, all_predictions)

# Print the evaluation metrics
print("\nPyTorch Model Evaluation:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")

# Print the classification report
print("\nClassification Report (PyTorch Model):")
print(classification_report(all_labels, all_predictions))
```
